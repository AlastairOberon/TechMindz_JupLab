```python
label1 = LabelEncoder()
label2 = LabelEncoder()
label3 = LabelEncoder()
```

```python
penguins_cleaned["species"] = label1.fit_transform(penguins_cleaned["species"])
penguins_cleaned["island"] = label2.fit_transform(penguins_cleaned["island"])
penguins_cleaned["sex"] = label3.fit_transform(penguins_cleaned["sex"])
```

```python
y = penguins_cleaned["species"]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 1)
```
