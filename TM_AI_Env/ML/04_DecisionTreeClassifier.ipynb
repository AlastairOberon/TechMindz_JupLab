```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
data1 = pd.read_csv(
    "/home/alastair_oberon/Shenanigans/Coding_shenanigans/Juplab_env/TM_AI_Env/Test_CSV/data_edit.csv"
)
data1
```

```python
data1.drop(["Unnamed: 0"], axis = 1, inplace = True)
data1
```

```python
ob1 = LabelEncoder()
ob2 = LabelEncoder()
```

```python
data1["country"] = ob1.fit_transform(data1["country"])
data1["Gender"] = ob2.fit_transform(data1["Gender"])
data1
```

```python
x_data1 = data1.drop(["purchase"], axis = 1)
y_data1 = data1["purchase"]
```

```python
x_data1_train,x_data1_test,y_data1_train,y_data1_test = train_test_split(x_data1,y_data1, test_size=0.3, random_state=1)
```

```python
model_data1 = DecisionTreeClassifier()
model_data1.fit(x_data1_train,y_data1_train)
```

```python
print(model_data1.score(x_data1_train, y_data1_train) * 100)
print(model_data1.score(x_data1_test, y_data1_test) * 100)
```

```python
pred_data1 = model_data1.predict(x_data1_test)
pred_data1
```

```python
y_data1_test
```

```python
cm = confusion_matrix(y_data1_test,pred_data1)
cm
```

```python
cr = classification_report(y_data1_test,pred_data1)

for i in cr:
    print(i,end = "")
```

```python
data2 = pd.read_csv(
    "/home/alastair_oberon/Shenanigans/Coding_shenanigans/Juplab_env/TM_AI_Env/Test_CSV/salaries.csv"
)
data2
```

```python
data2_enc_com = LabelEncoder()
data2_enc_job = LabelEncoder()
data2_enc_deg = LabelEncoder()
data2_enc_sal = LabelEncoder()
```
