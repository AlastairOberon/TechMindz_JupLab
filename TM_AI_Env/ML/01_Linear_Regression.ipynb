```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(
    "/data/Coding_Shennanigans/TechMindz/TM_AI/JupLab_env/TM_AI_Env/Test_CSV/homeprices.csv"
)
data
```

```python
x = data[["area"]]
y = data.price
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict([[0]])
```

```python
model.predict([[2700]])
```

```python
model.score(x, y) * 100
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
model.predict([[2800], [3000]])
```

```python
pre = model.predict(data[["area"]])
pre
```

```python
plt.xlabel("area in sqft")
plt.ylabel("prices in us$")
plt.scatter(data.area, data.price, marker="o", color="red")
plt.plot(data.area, pre, color="black")
plt.show()
```

```python
label_cop = LabelEncoder()
label_mod = LabelEncoder()
label_loc = LabelEncoder()
label_mil = LabelEncoder()
label_eng = LabelEncoder()
label_col = LabelEncoder()
label_asm = LabelEncoder()
label_bdy = LabelEncoder()
label_trn = LabelEncoder()
label_reg = LabelEncoder()
```

```python
pakwheels["Company Name"] = label_cop.fit_transform(pakwheels["Company Name"])
pakwheels["Model Name"] = label_mod.fit_transform(pakwheels["Model Name"])
pakwheels["Location"] = label_loc.fit_transform(pakwheels["Location"])
pakwheels["Mileage"] = label_mil.fit_transform(pakwheels["Mileage"])
pakwheels["Engine Type"] = label_eng.fit_transform(pakwheels["Engine Type"])
pakwheels["Color"] = label_col.fit_transform(pakwheels["Color"])
pakwheels["Assembly"] = label_asm.fit_transform(pakwheels["Assembly"])
pakwheels["Body Type"] = label_bdy.fit_transform(pakwheels["Body Type"])
pakwheels["Transmission Type"] = label_trn.fit_transform(pakwheels["Transmission Type"])
pakwheels["Registration Status"] = label_reg.fit_transform(
    pakwheels["Registration Status"]
)
```

```python
pakwheels
```

```python
pakwheels.corr()
```

```python
x = pakwheels.drop(["Price"], axis=1)
y = pakwheels["Price"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
print(model.score(xtrain, ytrain) * 100)
print(model.score(xtest, ytest) * 100)
```
