```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import statsmodels.formula.api as sms
import scipy.stats as st
```

```python
import_file = pd.read_excel(r"C:\Users\User\Desktop\FIN Tasks For Assignment.xlsx",sheet_name="Task 1 - Sample")
import_file
```

```python
age = pd.DataFrame(import_file, columns = ["age_range"])
age
```

```python
l_limit = pd.to_numeric(age["age_range"].str[:2], errors = "coerce")
l_limit
```

```python
u_limit = pd.to_numeric(age["age_range"].str[-2:], errors = "coerce")
u_limit
```

```python
age_range = pd.concat([l_limit,u_limit], axis = 1)
age_range.columns = ["l_limit", "u_limit"]
age_range["mid_point"] = (age_range["l_limit"] + age_range["u_limit"])/2
age_range.loc[0,"mid_point"] = 72
age_range
```

```python
temp = pd.to_numeric(import_file["Tot_inc_amt"])
temp.columns = ["Tot_inc_amt"]
temp
```

```python
tsk_5 = pd.concat([age_range, temp], axis = 1)
tsk_5
#"mid_point""Tot_inc_amt"
```

```python
print ('The correlation matrix is')
print(tsk_5[['mid_point', 'Tot_inc_amt']].corr())

print(sms.ols('Tot_inc_amt ~ mid_point', tsk_5).fit().summary())

sns.relplot(x = "mid_point", y = "Tot_inc_amt", data = tsk_5, height = 5, aspect = 2.5)
tsk_5 = tsk_5.dropna(subset=['mid_point'])
plt.xticks(tsk_5['mid_point'])
plt.ticklabel_format(style = "plain", axis = "y")
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Tot_inc_amt")
model = st.linregress (tsk_5 ['mid_point'], tsk_5 ['Tot_inc_amt'])
print(model)

m=model [0]
c= model [1]
x= tsk_5 ['mid_point']
plt.plot(x, m*x+c,'r')

plt.show()
```
