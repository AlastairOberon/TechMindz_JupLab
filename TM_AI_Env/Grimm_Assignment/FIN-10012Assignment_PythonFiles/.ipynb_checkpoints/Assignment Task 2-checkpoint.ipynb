```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
import_file = pd.read_excel(r"C:\Users\User\Desktop\FIN Tasks For Assignment.xlsx", sheet_name="Task 1 - Sample")
import_file
```

```python
sns.set(style ='whitegrid')
sns.countplot(y=import_file['Occ_code'], order=import_file['Occ_code'].value_counts(ascending=True).index).set(title="Occupation code")
plt.show()
```

```python
import_file['Occ_code'].value_counts()
```

```python
m = import_file[import_file.Gender =="Male"]["Gender"].count()
f = import_file[import_file.Gender =="Female"]["Gender"].count()
print("Male is", m)
print("Female is", f)
print("the total is", m+f)

sizes = [m,f]
labels = ["Male","Female"]
plt.pie(sizes, labels = labels, autopct="%1.2f%%")
plt.legend()
plt.show()
```

```python
import_file['Gender'].value_counts()
```
