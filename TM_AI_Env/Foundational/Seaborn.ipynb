```python
tips = sns.load_dataset("tips")
sns.barplot(
    x="day",
    y="tip",
    hue="sex",
    data=tips,
)

plt.xlabel("Day")
plt.ylabel("Average Tip Amount")
plt.title("Average Tip amount by Day")
```

```python
avg_tip_male_thursday = tips[(tips["day"] == "Thur") & (tips["sex"] == "Male")][
    "tip"
].mean()

print(f"Average tip amount given by men on Thursday: {avg_tip_male_thursday:.2f}")
```

```python
avg_tip_female_thursday = tips[(tips["day"] == "Thur") & (tips["sex"] == "Female")][
    "tip"
].mean()

print(f"Average tip amount given by men on Thursday: {avg_tip_female_thursday:.2f}")
```

```python
avg_tip_male_friday = tips[(tips["day"] == "Fri") & (tips["sex"] == "Male")][
    "tip"
].mean()

print(f"Average tip amount given by men on Thursday: {avg_tip_male_friday:.2f}")
```
