```python
x_range = np.arange(6)
width = 0.3

y = np.random.randint(50, 90, 6)
z = np.random.randint(50, 90, 6)
w = np.random.randint(50, 90, 6)
x_labels = ["Python", "R", "ML", "AI", "DS", "stats"]

plt.bar(x_range - width, y, width=width, label="Current Weight")
plt.bar(x_range, z, width=width, label="Weight Needed")
plt.bar(x_range + width, w, width=width, label="Ideal Weight")

plt.xticks(x_range, x_labels)

plt.title("Person Weight Comparison")
plt.xlabel("Category")
plt.ylabel("Weight")
plt.grid(True)
plt.legend()
plt.show()
```
